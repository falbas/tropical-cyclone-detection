```python
import xarray as xr
```

```python
file = 'http://182.16.248.173:8080/dods/INA-NWP/2023/2023111412/2023111412-d01-asim'
ds = xr.open_dataset(file)
```

```python
tc = ds['tc'][0][0]
rh = ds['rh'][0][0]
wspd = ds['wspd'][0][0]
u = ds['u'][0][0]
v = ds['v'][0][0]
```

```python
nc = xr.merge([tc,rh,wspd,u,v])
```

```python
nc.to_netcdf('sample/sample.nc')
```
