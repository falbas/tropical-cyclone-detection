```python
times = ds['time'].values
```

```python
dates_index = []
for date in dates:
  for i in range(0, len(times)):
    if date in str(times[i]):
      dates_index.append(i)
      print(f"{i} - {str(times[i])}")
      break
```

```python
lat = ds['latitude']
lon = ds['longitude']
```

```python
for d in dates_index:
  expver = 0 # ERA5 dataset version
  level = 2
  for i in range(d, d+8):
    # check dataset version
    if np.isnan(ds['u'][i][expver][level][0][0].values):
      expver = 1
      
    u = ds['u'][i][expver][level]
    v = ds['v'][i][expver][level]
    ws = np.sqrt(u**2 + v**2)

    plt.ioff() # disable interactive mode
    dpi = 100
    zoom = 3
    fig = plt.figure(figsize=(len(lon)*zoom/dpi, len(lat)*zoom/dpi))

    m = Basemap(projection='cyl', llcrnrlon=lon[0].values, llcrnrlat=lat[-1].values, urcrnrlon=lon[-1].values, urcrnrlat=lat[0].values, resolution='i')
    m.drawcoastlines(color='white')

    # cf = plt.contourf(lon, lat, ws, cmap='jet')

    skip = 3
    q = m.quiver(lon[::skip], lat[::skip], u[::skip,::skip], v[::skip,::skip], scale_units='xy', scale=5, width=0.0015)

    plt.tight_layout(pad=0)

    name = str(ds['time'][i].values).replace(':',".")+"-"+str(ds['level'][level].values)
    fig.savefig(f"output/10/1000/{name}.jpg", dpi=dpi)
```
