```python
from herbie import FastHerbie
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
import datetime
```

```python
dt = datetime.datetime.now()
date = dt.strftime("%Y-%m-%d")
# date = '2024-05-20'
```

```python
FH = FastHerbie(
    DATES=[date],
    model="ifs",
    product="oper",
    fxx=range(0, 147, 3),
)
```

```python
FH.objects
```

```python
search = ":[u|v]:1000:"
```

```python
FH.inventory(search)
```

```python
FH.download(search)
```

```python
ds = FH.xarray(search)
```

```python
ds
```

```python
valid_time = ds['valid_time']
lat = ds['latitude'][280:441]
lon = ds['longitude'][1040:1361]
```

```python
str(valid_time[0].values).split(':')[0]
```

```python
for i in range(0, len(valid_time)):
    u = ds["u"][i][280:441, 1040:1361]
    v = ds["v"][i][280:441, 1040:1361]
    ws = np.sqrt(u**2 + v**2)

    plt.ioff()  # disable interactive mode
    dpi = 100
    zoom = 3
    fig = plt.figure(figsize=(len(lon) * zoom / dpi, len(lat) * zoom / dpi))

    m = Basemap(
        projection="cyl",
        llcrnrlon=lon[0].values,
        llcrnrlat=lat[-1].values,
        urcrnrlon=lon[-1].values,
        urcrnrlat=lat[0].values,
        resolution="i",
    )
    m.drawcoastlines(color="white")

    # cf = plt.contourf(lon, lat, ws, cmap='jet')

    skip = 3
    q = m.quiver(
        lon[::skip],
        lat[::skip],
        u[::skip, ::skip],
        v[::skip, ::skip],
        scale_units="xy",
        scale=5,
        width=0.0015,
    )

    plt.tight_layout(pad=0)

    name = str(valid_time[i].values).replace(':',".")
    fig.savefig(f"output/12/{date}/{name}.jpg", dpi=dpi)
```
