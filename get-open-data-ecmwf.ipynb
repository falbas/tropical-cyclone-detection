```python
import ecmwf.data as ecdata
from magpye import GeoMap
from ecmwf.opendata import Client
import xarray as xr
import datetime
import numpy as np
```

```python
client = Client("ecmwf")
```

```python
dt = datetime.datetime.now()

# date = dt.strftime('%Y%m%d')
date = '20240514'
parameters = ['u', 'v']
filename = f'open-data/{date}.grib'
# step=[0, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33, 36, 39, 42, 45, 48, 51, 54, 57, 60, 63, 66, 69]
step=0

print(filename)
```

```python
client.retrieve(
    date=date,
    time=0,
    step=step,
    stream="oper",
    levelist=[1000],
    param=parameters,
    target=filename
)
```

```python
ds = xr.open_dataset(filename)
```
